```python
import pandas as pd
import sqlite3
from func_auxiliares.config import DB_PATH
from func_auxiliares.graficos_utils import get_df
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf

df=get_df("select año, importaciones from balanza_comercial", str(DB_PATH))
y=np.array(df['importaciones'], dtype=float)

# 1) Autocorrelación (lag 1, y también todos los lags hasta 20)
rho1 = np.corrcoef(y[1:], y[:-1])[0, 1]
r = acf(y, nlags=20, fft=False)
print("rho1 =", rho1)
print("ACF (lags 0..20) =", r)

# 2) Correlograma (ACF)
plot_acf(y, lags=20, alpha=0.05)
plt.show()
```
