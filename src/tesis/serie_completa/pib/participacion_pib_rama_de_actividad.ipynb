```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3, os
import numpy as np
#  1. Parmetros y rutas 
CRISIS_52_55     = slice(1952, 1955)   # Crisis
EXPANSION_56_69  = slice(1956, 1969)   # Expansin
RECESION_70_81   = slice(1970, 1981)   # Recesin
CRISIS_82_85     = slice(1982, 1985)   # Crisis (neoliberal)
EXPANSION_86_99  = slice(1986, 1999)   # Expansin
CRISIS_00_05     = slice(2000, 2005)   # Crisis
ACUMULACION_06_13= slice(2006, 2013)   # Expansin (acumulacin)
RECESION_14_23   = slice(2014, 2024)   # Recesin

OUTPUT_DIR = "../../../../assets/tesis/serie_completa"
os.makedirs(OUTPUT_DIR, exist_ok=True)

plt.rcParams.update({
    "font.family": "serif", "font.size": 12,
    "axes.titlesize": 16, "axes.labelsize": 14,
    "grid.linestyle": "--", "lines.linewidth": 2,
    "figure.dpi": 150, "savefig.bbox": "tight"
})

#  2. Carga de datos 
with sqlite3.connect("../../../../db/proyectomacro.db") as conn:
    df = pd.read_sql_query("SELECT * FROM participacion_pib_ramas", conn)

df.set_index("ao", inplace=True)

# Columnas a graficar (excluimos minas_canteras_total)
cols = [
    "agropecuario",
    "mineria",
    "petroleo_crudo_y_gas_natural",
    "industria_manufacturera",
    "construcciones",
    "energia",
    "transportes",
    "comercio_finanzas",
    "gobierno_general",
    "propiedad_vivienda",
    "servicios"
]
pct = df[cols]

#  3. Estadsticas promedio por periodo 
avg_crisis_52_55      = pct.loc[CRISIS_52_55].mean()
avg_expansion_56_69   = pct.loc[EXPANSION_56_69].mean()
avg_recesion_70_81    = pct.loc[RECESION_70_81].mean()
avg_crisis_82_85      = pct.loc[CRISIS_82_85].mean()
avg_expansion_86_99   = pct.loc[EXPANSION_86_99].mean()
avg_crisis_00_05      = pct.loc[CRISIS_00_05].mean()
avg_acumulacion_06_13 = pct.loc[ACUMULACION_06_13].mean()
avg_recesion_14_23    = pct.loc[RECESION_14_23].mean()
#offset
hitos_offset = {
    1951: (1, 0.10),
    1955: (3, 0.05),
    1969: (3, 0.05),
    1981: (1, 0.10),
    1985: (3, 0.05),
    1999: (1, 0.10),
    2005: (3, 0.05),
    2014: (3, 0.05)
}

#  4. Grfico stacked-bar 
fig, ax = plt.subplots(figsize=(14, 7))
pct.plot(kind="bar", stacked=True, ax=ax, width=0.8)

ax.set_ylabel("Participacin (%)")
ax.set_xlabel("Ao")
ax.set_title("Participacin sectorial del PIB 1950-2023",
             fontweight="bold")

# muevo la leyenda fuera del rea de grfico, abajo
ax.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.15),
    ncol=6,            # ajusta segn cuntos tems tengas
    fontsize=10,
    frameon=False
)
fig.subplots_adjust(bottom=0.25)   # expande margen inferior

# construyo un array 0,1,2, tan largo como df.index
positions = np.arange(len(df.index))
# uso solo cada segunda posicin, y como etiqueta el ao correspondiente
plt.xticks(positions[::2], df.index[::2], rotation=45)

#  4.5 Lneas verticales y texto de hitos 
hitos_v = {
    1951: "Crisis",
    1955: "Expansin",
    1969: "Recesin",
    1981: "Crisis",
    1985: "Expansin",
    1999: "Crisis",
    2005: "Expansin",
    2014: "Recesin"
}
fig.subplots_adjust(right=0.72)

for yr, lbl in hitos_v.items():
    if yr in pct.index:
        idx = pct.index.get_loc(yr)
        y_max=ax.get_ylim()[1]
        ax.axvline(
            x=idx + 0.5, color="gray", ls="--", lw=1.5, zorder=5
        )
        dx,dy=hitos_offset.get(yr,(0,0.5))
        ax.text(
            idx + 0.5+dx,dy, lbl, rotation=0 if lbl=='Expansin' or lbl=='Recesin' else 90, ha="center", va="top",
            fontsize=12, color="black",
            transform=ax.get_xaxis_transform(), clip_on=False,
            bbox=dict(facecolor="white", alpha=0.85, edgecolor="none"),
            zorder=6
        )

#  6. Guardar y mostrar 
plt.tight_layout()
out_path = os.path.join(OUTPUT_DIR, "participacion_pib_ramas.png")
plt.savefig(out_path, dpi=300)
plt.show()

out_path
```

```python
import seaborn as sns
df_means = pd.DataFrame({
    #"Crisis 5255":      avg_crisis_52_55,
    "Expansin 5668":   avg_expansion_56_69,
    "Recesin 7081":    avg_recesion_70_81,
    #"Crisis 8285":      avg_crisis_82_85,
    "Expansin 8699":   avg_expansion_86_99,
    #"Crisis 0005":      avg_crisis_00_05,
    "Acumulacin 0613": avg_acumulacion_06_13,
    "Recesin 1423":    avg_recesion_14_23
}).round(1)

#  5. Grfico seaborn (tabla en grises sobrio) 
fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(
    df_means,
    annot=True, fmt=".1f",
    cmap="Blues",
    cbar=False,
    linewidths=0.5,
    linecolor="lightgray",
    ax=ax
)

ax.set_title(
    "Promedios de participacin sectorial del PIB\npor ciclo econmico (19522023)",
    fontweight='bold', pad=20
)
ax.set_ylabel("Rama de actividad")
ax.set_xlabel("Periodo econmico")
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
out_file = os.path.join(OUTPUT_DIR, "tabla_promedios_ciclos_seaborn.png")
fig.savefig(out_file)
plt.show()
```
